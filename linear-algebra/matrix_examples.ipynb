```julia
using DSP
a = [1, 0, -1]
b = [2, 1, -1]
c = conv(a, b)
```

```julia
u = rand([-1, 1], 50)
c = [1, 0.7, -0.3]
y = conv(u, c)
```

```julia
using Plots 

h = [.9, -.5, .5, -.4, .3, -.3, .2, -.1]

z = conv(h, y)

plot(u, label="original signal", color="black")
plot!(y, label="received signal", color = "green")
plot!(z, label="equalized received signal", color="blue")
```
